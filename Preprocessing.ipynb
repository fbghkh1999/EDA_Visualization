```python
# part2
imp=SimpleImputer(missing_values=  np.nan, strategy= 'median')
imp.fit(dataset)
dataset=imp.transform(dataset)
dataset=pd.DataFrame(dataset)

dataset=dataset.rename(columns={0:'t',1:'E_Plug',2:'E_Heat',3:'Price',4:'Temperature',5:'Occupants'})
```

```python
"""
find outlier
"""
dataset.boxplot()
# 
a=dataset['E_Plug'].quantile(0.25)
b=dataset['E_Plug'].quantile(0.75)
IQR=b-a
#MILD OUTLIER
lower=a-1.5*IQR
lower
higher=b+1.5*IQR
higher
dataset.replace(120,42,inplace=True)
dataset
```

```python
# dataset.boxplot()
x=dataset['E_Heat'].quantile(0.25)
y=dataset['E_Heat'].quantile(0.75)
IQRMID=y-x
lowermid=x-1.5*IQRMID
lowermid
sns.boxplot(data=dataset)
dataset['E_Heat'][dataset['E_Heat']<=lowermid]
dataset.replace(-4,16,inplace=True)
```

```python
#CONCAT data
dataset_new=pd.read_csv('Data_New.csv')
dataset_new
```

```python
dataset=pd.concat([dataset,dataset_new],axis=1)
```

```python
dataset=pd.get_dummies(dataset)
dataset
```
